```python
#importing libraries
import pandas as pd
from dateutil import parser
```

```python
# Read the CSV file 
df = pd.read_csv('statsfinal.csv')
```

```python
# Define a custom function to parse dates
def parse_date(date_str):
    try:
        return parser.parse(date_str)
    except:
        return None
```

```python
# Apply the custom date parsing function to the 'Date' column
df['Date'] = df['Date'].apply(parse_date)
# Drop rows with invalid dates
df = df.dropna(subset=['Date'])
df
```

```python

# Extract year, month, and day from the parsed date
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.strftime('%b')  # Abbreviated month names
df['Day'] = df['Date'].dt.day
df
```

```python
# Derive weekday from the 'Date' column and add it as a new column 'Weekday'
df['Weekday'] = df['Date'].dt.day_name()
df
```

```python
# Calculate the total sales and total revenue
df['Total Sales'] = df['Q-P1'] + df['Q-P2'] + df['Q-P3'] + df['Q-P4']
df['Total Revenue'] = df['S-P1'] + df['S-P2'] + df['S-P3'] + df['S-P4']
df
```

```python
# Rename columns
df = df.rename(columns={
    'Q-P1': 'Q1',
    'Q-P2': 'Q2',
    'Q-P3': 'Q3',
    'Q-P4': 'Q4',
    'S-P1': 'S1',
    'S-P2': 'S2',
    'S-P3': 'S3',
    'S-P4': 'S4'
})
df
```

```python
# Reorder the columns
df = df[['Date', 'Day', 'Month', 'Year', 'Weekday', 'Q1', 'Q2', 'Q3', 'Q4', 'Total Sales', 'S1', 'S2', 'S3', 'S4', 'Total Revenue']]
df
```

```python
# Save the preprocessed data to a CSV file named 'PreprocessedData.csv' without including the index
df.to_csv('PreprocessedData.csv', index=False)
```
